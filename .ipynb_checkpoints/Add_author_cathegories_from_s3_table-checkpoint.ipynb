```python
import pandas as pd
import json

# Define path to the JSON file
json_file_path = './intermediatte_files/Siletti_from_abc.json'

# Load the JSON file
with open(json_file_path, 'r') as file:
    json_data = json.load(file)

# Convert JSON to DataFrame
# Assuming you want to convert the 'annotations' part of the JSON
annotations_df = pd.json_normalize(json_data['annotations'])

# Display the DataFrame
annotations_df.head()  # Display the first few rows of the DataFrame
```

```python
import pandas as pd

# Define paths to the files
csv_file_path = './Author_cathegories/science.add7046_table_s3.csv'

# Load the CSV file
csv_df = pd.read_csv(csv_file_path)

# Assuming annotations_df is already created from the JSON file
# Merge the two DataFrames on the 'cell_label' column
merged_df = pd.merge(annotations_df, csv_df, on='cell_label', how='left')

# Display the first few rows of the merged DataFrame
merged_df.head()

# Optionally, save the merged DataFrame to a new CSV or JSON file
merged_df.to_csv('./intermediatte_files/Siletti_added_author_cathegories.csv', index=False)
merged_df.to_json('./intermediatte_files/Siletti_added_author_cathegories.json', orient='records', indent=4)

print("Merged DataFrame saved to './intermediatte_files/Siletti_added_author_cathegories.csv' and '.json'") 
#this notebook can be modified too add less author cathegories to avoid superfluous informations
```
