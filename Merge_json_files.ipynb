```python
import json
from deepmerge import Merger

# Define the custom merge strategy for dictionaries
def custom_dict_strategy(config, path, base, nxt):
    for key in nxt:
        if key == "labelset":
            # If key is "labelset", keep the value from the second dictionary
            base[key] = nxt[key]
        elif key in base:
            # If key exists in both dictionaries and is not "labelset", append the values
            if isinstance(base[key], list) and isinstance(nxt[key], list):
                base[key].extend(nxt[key])
            else:
                base[key] = [base[key], nxt[key]] if not isinstance(base[key], list) else base[key] + [nxt[key]]
        else:
            base[key] = nxt[key]
    return base

# Define the custom merge strategy for lists
def custom_list_strategy(config, path, base, nxt):
    return base + nxt

# Create a merger object with the custom strategies
merger = Merger(
    [(list, custom_list_strategy), (dict, custom_dict_strategy)],
    ["override"],
    ["override"]
)

# Load the JSON files
with open('./Final_json_files/Siletti_added_author_categories_v2_neurons_copy.json', 'r') as f1:
    data1 = json.load(f1)

with open('./Final_json_files/Siletti_added_author_categories_v2_nn.json', 'r') as f2:
    data2 = json.load(f2)

# Merge the data using the custom strategy
merged_data = merger.merge(data1, data2)

# Save the merged data to a new JSON file
merged_file_path = './Final_json_files/merged_clusters.json'
with open(merged_file_path, 'w') as f:
    json.dump(merged_data, f, indent=2)

print(f"Merged data saved to {merged_file_path}")
```
